```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_data_AMRITESH UPADHYAYA.csv', encoding= 'unicode_escape')
```

```python
%matplotlib inline
```

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Create subplots for multiple histograms
fig, axes = plt.subplots(2, 2, figsize=(12, 10), constrained_layout=True)  # Added constrained_layout

# Flatten axes to avoid indexing issues
axes = axes.flatten()

# List of columns to plot
columns = ["Age", "BMI", "MentalHealth", "PhysicalHealth"]
titles = ["Distribution of Age", "Distribution of BMI", "Distribution of Mental Health Days", "Distribution of Physical Health Days"]

# Loop through and plot
for i, col in enumerate(columns):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(titles[i])

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the plotting style
sns.set(style="whitegrid")

# Create subplots for categorical variables
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# List of categorical columns (Fixed column name)
columns = ["HighBP", "HighCholestrol", "Stroke"]  # Updated "HighChol" to "HighCholestrol"
titles = ["Prevalence of High Blood Pressure", "Prevalence of High Cholesterol", "Prevalence of Stroke"]

# Loop through and plot
for i, col in enumerate(columns):
    if col in df.columns:  # Ensure the column exists
        sns.countplot(x=df[col], hue=df[col], ax=axes[i], palette="coolwarm", legend=False)
        axes[i].set_title(titles[i])
        axes[i].set_xticks([0, 1])  # Fix xticks
        axes[i].set_xticklabels(["No", "Yes"])
    else:
        print(f"⚠️ Warning: Column '{col}' not found in DataFrame!")

plt.show()
```

```python
# Count the number of cases for heart disease
heart_disease_counts = df["HeartDiseaseorAttack"].value_counts()

# Plot Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(heart_disease_counts, labels=["No Heart Disease", "Heart Disease"], 
        autopct='%1.1f%%', colors=["lightblue", "red"], startangle=90)
plt.title("Distribution of Heart Disease Cases")
plt.show()
```

```python
# Define categorical variables and titles
cat_columns = ["HighBP", "HighCholestrol"]
titles = ["Heart Disease vs High Blood Pressure", "Heart Disease vs High Cholesterol"]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Loop through and create bar plots
for i, col in enumerate(cat_columns):
    sns.countplot(x=col, hue="HeartDiseaseorAttack", data=df, ax=axes[i], palette="coolwarm")
    axes[i].set_title(titles[i])
    axes[i].set_xticks([0, 1])  
    axes[i].set_xticklabels(["No", "Yes"])  

plt.tight_layout()
plt.show()

# Box Plot for BMI Distribution
plt.figure(figsize=(8, 6))
sns.boxplot(x="HeartDiseaseorAttack", y="BMI", data=df, palette="coolwarm")
plt.xticks([0, 1], ["No Heart Disease", "Heart Disease"])
plt.title("BMI Distribution by Heart Disease Status")
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = df.corr()

# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Variables")
plt.show()

# Scatter Plot: Age vs BMI (Colored by Heart Disease)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Age", y="BMI", hue="HeartDiseaseorAttack", data=df, alpha=0.7, palette="coolwarm")
plt.title("Age vs BMI (Colored by Heart Disease Status)")
plt.show()
```

```python
# Create age groups
df["AgeGroup"] = pd.cut(df["Age"], bins=[18, 35, 50, 65, 80, 100], 
                        labels=["18-35", "36-50", "51-65", "66-80", "81+"])


# Plot heart disease prevalence by age group
plt.figure(figsize=(8, 6))
sns.countplot(x="AgeGroup", hue="HeartDiseaseorAttack", data=df, palette="coolwarm")
plt.title("Heart Disease Prevalence Across Age Groups")
plt.show()

# Heart Disease by Gender
plt.figure(figsize=(6, 5))
sns.countplot(x="Sex", hue="HeartDiseaseorAttack", data=df, palette="coolwarm")
plt.title("Heart Disease Distribution Across Gender")
plt.show()

# Heart Disease by Education Level
plt.figure(figsize=(8, 6))
sns.countplot(x="Education", hue="HeartDiseaseorAttack", data=df, palette="coolwarm")
plt.title("Heart Disease Prevalence Across Education Levels")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "cleaned_data_AMRITESH UPADHYAYA.csv"
df = pd.read_csv(file_path)
```

```python
# Set the style
sns.set_theme(style="whitegrid")

# List of categorical variables
categorical_vars = ["Smoker", "PhysicalActivity", "Fruits", "Veggies"]

# Create bar plots for categorical variable distribution
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for ax, var in zip(axes.flatten(), categorical_vars):
    sns.countplot(x=df[var], ax=ax)  # Removed 'palette' to fix the warning
    ax.set_title(f"Distribution of {var}")
    ax.set_xticks([0, 1])  # Ensures fixed ticks before setting labels
    ax.set_xticklabels(["No", "Yes"])

plt.tight_layout()
plt.show()
```

```python
# Create grouped bar plots for heart disease prevalence across categories
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for ax, var in zip(axes.flatten(), categorical_vars):
    sns.countplot(x=df[var], hue=df["HeartDiseaseorAttack"], ax=ax, palette="coolwarm")
    ax.set_title(f"Heart Disease Prevalence by {var}")
    ax.set_xticks([0, 1])  # Fix tick issue
    ax.set_xticklabels(["No", "Yes"])
    ax.set_ylabel("Count")
    ax.legend(title="Heart Disease", labels=["No", "Yes"])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "cleaned_data_AMRITESH UPADHYAYA.csv"
df = pd.read_csv(file_path)
```

```python
# Set style
sns.set_theme(style="whitegrid")

# List of continuous variables
continuous_vars = ["BMI", "MentalHealth", "PhysicalHealth"]

# Create histograms
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, var in zip(axes, continuous_vars):
    sns.histplot(df[var], bins=30, kde=True, ax=ax)
    ax.set_title(f"Histogram of {var}")

plt.tight_layout()
plt.show()

# Create box plots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, var in zip(axes, continuous_vars):
    sns.boxplot(y=df[var], ax=ax)
    ax.set_title(f"Box Plot of {var}")

plt.tight_layout()
plt.show()
```

```python
# Compute the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Define variables of interest
scatter_vars = ["BMI", "MentalHealth", "PhysicalHealth"]

# Create scatter plots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, var in zip(axes, scatter_vars):
    sns.scatterplot(x=df[var], y=df["HeartDiseaseorAttack"], alpha=0.5, ax=ax)
    ax.set_title(f"{var} vs. Heart Disease")

plt.tight_layout()
plt.show()
```

```python
# Assuming an 'Age' column exists
df["AgeGroup"] = pd.qcut(df["BMI"], q=5, labels=["Very Low", "Low", "Medium", "High", "Very High"])

# Plot line chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="AgeGroup", y="BMI", estimator="mean", ci=None, marker="o")
plt.title("BMI Trends Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Average BMI")
plt.show()
```

```python
# Create violin plots for continuous variables
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, var in zip(axes, continuous_vars):
    sns.violinplot(y=df[var], ax=ax)
    ax.set_title(f"Violin Plot of {var}")

plt.tight_layout()
plt.show()
```

```python
# Select variables for pair plot
pair_vars = ["BMI", "MentalHealth", "PhysicalHealth", "HeartDiseaseorAttack"]

# Create pair plot
sns.pairplot(df[pair_vars], hue="HeartDiseaseorAttack", palette="coolwarm")
plt.show()
```
